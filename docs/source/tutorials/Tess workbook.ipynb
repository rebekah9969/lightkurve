# An Introduction to TESS and  its data products:

The Transiting Exoplanet Survey Satellite (TESS) was launched April 18th 2018 with the scientific objective of finding plants outside of our own solar system, including those that could support life. TESS is achieving scientific goal through the observation  of hundreds of thousands of brightest stars near near our own Sun. 

TESS's orbit is elliptical and has a period of 13.7 days, with a nominal perigee and apogee of 17
R$_{Earth}$ and 59 R$_{Earth}$, respectively.

TESS has surveyed the entire sky over the course of  its two year primary mission. 


The data collected by TESS is housed by the Barbara A. Mikulski Archive for Space Telescopes (MAST) at the Space Telescope Science Institute (STScI). This MAST portal houses data for a number of telescopes, allowing a user to easily access all data for a given field or object of interest. For TESS the data itself its handled by NADA Goddard, with the TESS PI institute being MIT.  All exoplanet information obtained by TESS is stored by NExScI. A list of these various institutes and their affiliation with TESS are listed below.

## Instrument Overview

### Cameras and FoV:
The TESS instrument consists of 4 wide field of view cameras (FoV) with hoods, mount, sun-shield, and data handling unit (DHU). The cameras are refractive and act as a 1 x 4 array which map out a 24x96 degree FoV (or ~2300 deg$^{2}$). The four CCDs in each array are separated by 2mm and create an effective 4096 x 4096 pixel detector that is operated at -75 degrees C to reduce dark current. TESS's resolution is 21 arc-seconds per pixel.

Within its two year prime mission TESS broke the sky up into 13 northern and 13 southern hemisphere sectors. Each sector was observed for ~27.4 days with each hemisphere being observed for a year in total.

Explore a 3D version of the TESS setup [here](https://exoplanets.nasa.gov/alien-worlds/tess-spacecraft/)

### TESS Bandpass
The TESS bandpass is wide and red with a wavelength ranging from 600 - 1000 nm, centered on the Cousins-I band (I$_{C}$- 786.5 nm). This wide, red-optical bandpass is preferred as it reduces photon-counting noise and increases the sensitivity to small planets transiting cool, red stars.

## Data Products 

During science operations the four camera take a continuous stream of images each with a 2 second exposure. 

From these two second integrations there are several data products produced by the TESS mission, these include postage stamps (target pixel files (TPFs)) of selected targets and full-frame images (FFIs). Each of these products are created on-board and are co-adds of the two second exposures. Data products are provided in fits formats with multiple extensions which provide additional information on calibration, cosmic-ray rejection, and in the case of the TPFs aperture masks indicating the pixels used in [photometry](https://en.wikipedia.org/wiki/Photometry_(astronomy)#Absolute_photometry).

### Primary Data Products 


- **Target Pixel Files:**
    TESS TPFs contain all the pixels for a single cadence target, in one observing sector. If a target is observed in more than one sector, multiple TPFs will be created for that target. The images in the TPF have dimensions equal to the bounding box of the pixels that were collected for that target. To understand more about TPS and how to use the data please visit [this page](https://github.com/spacetelescope/notebooks/blob/master/notebooks/MAST/TESS/beginner_how_to_use_tp/beginner_how_to_use_tp.ipynb). There are two flavors of TPFs,
    - 20 second (2020 - Cycle 3 product) 
    - 2 minute (2018 onwards) 

    
- **Full Frame Images:**
  TESS FFIs contain all the pixels on a single CCD. A full TESS observing sector consists of 4 FFIs.
  FFI data will be provided in three forms uncalibrated, calibrated, and uncertainty. To understand more about FFIs and how to use the data please visit [this page](https://github.com/spacetelescope/notebooks/blob/master/notebooks/MAST/TESS/beginner_how_to_use_ffi/beginner_how_to_use_ffi.ipynb). There are two flavors of FFIs, although only one is currently being taken.
    - 10 minute (2020 - Cycle 3 product)
    - 30 minute (Cycles 1 & 2 only)

    
- [**Light Curves:**](https://imagine.gsfc.nasa.gov/science/toolbox/timing1.html)
    These contain the output of the photometric extraction and subsequent systematic removal. A single light curve file contains the data for one target in one observing sector. If a target is observed in more than one sector multiple light curve files will be created. The simple aperture photometry (SAP) flux series is known as the SAP_FLUX, while the flux series that has the common instrumental systematics removed using the Co-trending Basis Vector (CBV) files (see secondary data products) is known as the PDCSAP_FLUX. The PDCSAP_FLUX light curves also correct for the amount of flux captured by the photometric aperture and crowding from known nearby stars. To learn more about how to use light curve files please visit [this page](https://github.com/spacetelescope/notebooks/blob/master/notebooks/MAST/TESS/beginner_how_to_use_lc/beginner_how_to_use_lc.ipynb).

### Secondary Data Products
There are several secondary TESS data products, these include the following 

 - **Collateral Target Pixel Files:** These archive the pixel level calibration data collected per CCD, per camera, during science operations.
 
 - **Co-trending Basis Vectors (CBVs):** These represent the set of systematic trends present in the light curve data for each CCD.  These are useful as they can remove common instrumental effects from uncorrected light curve data if standard co-trended data is insufficient.
 
 - **Transit Signal Files:** These are files related to the automated search for transit signals.

## The TESS Data Flow

The flow of data from the space-craft to the archive involves several steps and institutions. The overall  flow can be seen in Figure 7, and is described below.

- Data is downloaded via the Deep Space Network and delivered to the Payload Operations Center (POC) at MIT. 
-  The POC sends uncalibrated re-quantized pixel data, target lists, spacecraft configuration and engineering data, and focal plane characterization models to the Science Processing Operations Center (SPOC) at NASA Ames.
- Using the instrument calibration models provided by the POC the SPOC calibrates the science data. The SPOC does this on the orbit first and then by the sector. When a sector is calibrated the transiting planet search software is run to identify and flag threshold crossing events (TCEs). 
- TCEs, calibrated TPFs, FFIs, and light curves generated from 2-min cadence targets are sent to the TESS Science Office (TSO). 
- The TSO conducts a detailed analysis of the TCEs and identifies TESS Objects of Interest (TOIs). 
- The TOIs are delivered to MAST.
- The processed data and meta-data from the SPOC are also archived at MAST.

## The TESS PSF
TESS was designed as a high precision photometer and not as an imager, as such the photometric stability and minimization of noise was more important than a compact point spread function (PSF).
The TESS PSF varies significantly across the focal plane due primarily to the optics. Instead of a PSF, TESS has a pixel response function (PRF), which represents the observed appearance of the point sources. 

The TESS PRF was created by the SPOC by fitting to micro-dithered data taken during PRF commissioning exercises. PRF models for sectors 1-3 can be found [here](https://archive.stsci.edu/missions/tess/models/prf_fitsfiles/start_s0001/), with sector 4 onwards [here](https://archive.stsci.edu/missions/tess/models/prf_fitsfiles/start_s0004/).

![tess_psf.png](images/tess_psf.png)
**Figure 8:** *PRF from Sector 1, Camera 1.*

Physical WCS solutions can be used to convert the PRF image coordinates into the corresponding TESS CCD's.
For more information about the TESS PRF see page 49 of the [instrument handbook](https://archive.stsci.edu/files/live/sites/mast/files/home/missions-and-data/active-missions/tess/_documents/TESS_Instrument_Handbook_v0.1.pdf).

Given the unusual nature of the TESS PRF, photometry of an object is typically obtained through the summation of all pixels within a given region. This region is referred to as an "aperture mask" and can be determined through the pipeline or can be selected  by the user,  unlike typical apertures TESS aperture is not circular. 

## TESS Systematics

Here we outline some of the major sources of systematic effects on the TESS data. Several of these effects may be corrected for within the data processing pipeline, but residual effect may still be present within the user data files, it is therefore important to identify these issue and try to resolve them. 

Additional information on all systematics effects not identified here can be found within sections 6, 7, and 8 of the [TESS Instrument Handbook](https://archive.stsci.edu/files/live/sites/mast/files/home/missions-and-data/active-missions/tess/_documents/TESS_Instrument_Handbook_v0.1.pdf).

    
**Cosmic Rays:**
Cosmic-ray hits on the TESS CCD's are a significant source of noise. For FFIs this is  especially true and nearly half of the pixels in the 30 min FFIs are affected. Within the DHU a tools was developed to help mitigate the effect  of  the Cosmic-rays, images are  stacked and pixels are examined in groups of N. The highest and lowest values of the stack are removed, and the remaining sum are used to create the stack. Although this method of cosmic-ray rejection reduces contamination by a factor of  100,  some  low level outliers  still exist and can be seen within the data. These outliers  can however be removed via [TESS-zap](https://github.com/zkbt/tess-zap).

**Scattered Light:**
Each camera has a lens hood to reduce the scattered light from the Earth and the Moon. Due to TESS's wide field of view and the physical restrictions of the Sun shade the lens hood is not 100% efficient. The effect  of the scattered  light on the CCD's can be seen in the video  below, typically the patchy brightness  is 2-6 times that of the nominal sky background and covers approximately 10-15% of the FoV. When the Earth is  below the level  of the sun shade there  is no scattered light. When the Earth or Moon is directly in the FoV of a camera the data is no longer viable.

[![scatter.png](images/scatter.png)](https://www.youtube.com/watch?v=SP4QSF9G6FA "Scattered Light")

**Pointing Jitter:**
A key source of systematic noise comes from random pointing variations i.e., spacecraft jitter. This causes changes in the measured brightness as stars move across CCD pixels. The floor for this intrinsic noise is assumed to be 60 ppm on hourly timescales.
